```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
%%time
data = pd.read_excel('indeed_results_pp_2020-04-25.xlsx')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
x = data[['Education_Tenth','Education_Twelvth','Education_Bachelors','Education_Masters','Education_Doctorate','XP_Experience']].values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler(feature_range=(0, 1))
y = sc.fit_transform(data['Salary_Average'].values.reshape(-1, 1))
```

```python
lin = LinearRegression()
```

```python
lin.fit(x,y)
```

```python
y_pred = lin.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Accuracy of LR',mean_squared_error(y_pred,y))
```

```python
acc = 1 - mean_squared_error(y_pred,y)
```
